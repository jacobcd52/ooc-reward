```python
from finetune import load_models, Config, train
```

```python
cfg = Config(
    verbose=True,
    lr=1e-3,
    num_epochs=10,
    batch_size=4,
    beta=0.1,
    max_cot_length=100,
    new_cot_freq=10,
    lora_alpha=16,
    lora_r=16,
    model_name="meta-llama/Llama-2-7b-chat-hf"
)

# Train the model
tokenizer, model_to_finetune, original_model = load_models(cfg)
tokenizer.pad_token = tokenizer.eos_token
train(model_to_finetune, original_model, tokenizer, cfg)
```
